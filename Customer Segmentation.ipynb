```python
plt.figure(figsize=(15,5))
sns.countplot(y='Gender',data=data)
plt.show()
```

```python
sns.barplot(x= 'Gender', y= 'Age', data=data)
plt.show()
```

```python
plt.figure(1,figsize=(25,15))
n=0
for a in ['Age' , 'Annual Income (k$)' , 'Spending Score (1-100)']:
    n+=1
    plt.subplot(1 , 3 , n)
    sns.violinplot(x= a , y= 'Gender' , data=data)
plt.show()
```

```python
age_18_25 = data.Age[(data.Age>=18) & (data.Age<=25)]
age_26_35 = data.Age[(data.Age>=26) & (data.Age<=35)]
age_36_45 = data.Age[(data.Age>=36) & (data.Age<=45)]
age_46_55 = data.Age[(data.Age>=46) & (data.Age<=55)]
age_55_above = data.Age[(data.Age>=56)]

plt.figure(figsize=(10,10))
agex = ["18-25","26-35","36-45","46-55","55+"]
agey = [len(age_18_25.values), len(age_26_35.values), len(age_36_45.values), len(age_46_55.values), len(age_55_above.values)]
sns.barplot(x=agex, y=agey)
plt.xlabel("Age")
plt.ylabel("Number of Customers")
plt.show()
```
